```python
#smoker-charges 상관계수가 가장 크므로 그 둘의 관계를 파악
#smoker는 범주형, charges는 연속형이므로 smoker의 범주별로 charges가 어떻게 분포하는지 파악
#2개의 그래프를 한 셀에 표현
import matplotlib.pyplot as pyplot
ax = plt.figure(figsize = (12, 5)) #그림을 그릴 칸을 생성, figsize는 그림 크기
ax_1 = ax.add_subplot(121) #인자는 각각 행, 열, 인덱스
sns.distplot(df.loc[df["smoker"]==1, "charges"], ax = ax_1) #distplot는 분포 그래프로 히스토그램에 선 얹어줌
ax_1.set_title("Distribution of charges of smokers")
ax_2 = ax.add_subplot(122)
sns.distplot(df.loc[df["smoker"]==0, "charges"], ax = ax_2)
ax_2.set_title("Distribution of charges of Non-smokers")
```

```python
#categorical-numerical 분석을 하는 다른 방법 : side-by-side boxplot
plt.boxplot([df.loc[df["smoker"]==1, "charges"], df.loc[df["smoker"]==0, "charges"]]) #단일 boxplot이라면 리스트로 씌울 필요 없음
plt.title("Boxplots of charges-smoker")
plt.xticks([1, 2], ["smoker", "non-smoker"])
plt.show()
```
